## Automação e Pipelines

```python
def read_data(filename):
  path = "/content/drive/MyDrive/Aulas - DS/dados/"
  return pd.read_csv(path + filename)

def basic_info(df):
  print("Formato do dataset (linhas, colunas):", df.shape)
  missing_values = df.isnull().sum()
  missing_values = missing_values[missing_values > 0]
  if missing_values.empty:
    print("Base com dados faltantes!")
  else:
    print("Base sem dados faltantes")
    print("Colunas com dados faltantes", missing_values.index)

def encoding_column(df, column_name):
  df_dummies_features = pd.get_dummies(df[column_name], dtype=int, prefix=column_name)
  return pd.concat([df.drop(column_name, axis=1), df_dummies_features], axis=1)

def transform_category_columns(df):
  columns = df.select_dtypes(include="object").columns
  for col in columns:
    df = encoding_column(df, col)
  return df

def normalize_column(df, column_name):
  df[column_name] = (df[column_name] - df[column_name].min()) / (df[column_name].max() - df[column_name].min())
  return df

def normalize_numeric_columns(df):
  columns = df.select_dtypes(include="number").columns
  for col in columns:
    df = normalize_column(df, col)
  return df

def process(filename):
  df = read_data(filename)
  basic_info(df)
  df = transform_category_columns(df)
  df = normalize_numeric_columns(df)

  return df
```

```python
process("gym_members_exercise_tracking.csv")
```
