### Temos Dados Duplicados?

```python
len(df)
```

```python
df_duplicated = df.groupby(df.columns.tolist(),as_index=False).size()
df_duplicated[df_duplicated["size"]>1]
```

```python
columns_to_use = [column for column in df.columns if column != "Unnamed: 0"]

df_dulicated = df.groupby(columns_to_use,as_index=False).size()
df_dulicated[df_dulicated["size"]>1]
```
